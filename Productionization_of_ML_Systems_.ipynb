**Merge Flights and Hotels**

````python
# Merge Flights and Hotels
#  MERGE + SAVE FINAL DATA


# âœ… Load datasets from your folder (travel_capstone)
flights = pd.read_csv("travel_capstone/flights.csv")
hotels = pd.read_csv("travel_capstone/hotels.csv")
users = pd.read_csv("travel_capstone/users.csv")

# Step 1: Merge flights and hotels
flights_hotels = pd.merge(flights, hotels, on="travelCode", how="inner")

# Step 2: Merge flights_hotels with users
travel_data = pd.merge(flights_hotels, users, left_on="userCode_x", right_on="code", how="inner")

# Step 3: Clean up columns
travel_data.rename(columns={'userCode_x': 'userCode'}, inplace=True)
if 'userCode_y' in travel_data.columns:
    travel_data.drop(columns=['userCode_y'], inplace=True)

# Step 4: Rename columns for clarity
travel_data.rename(columns={
    'price_x': 'flight_price',
    'price_y': 'hotel_price',
    'date_x': 'flight_date',
    'date_y': 'hotel_date',
    'name_x': 'hotel_name',
    'name_y': 'user_name'
}, inplace=True)

# Step 5: Drop redundant 'code' column
travel_data.drop(columns=['code'], inplace=True)

# Step 6: Save merged dataset for model training
travel_data.to_csv('travel_capstone/travel_data.csv', index=False)

print(" Final merged dataset created and saved as 'travel_capstone/travel_data.csv'")
print("Shape:", travel_data.shape)
travel_data.head()

````

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.


# 1. Convert date columns to datetime
travel_data['flight_date'] = pd.to_datetime(travel_data['flight_date'])
travel_data['hotel_date'] = pd.to_datetime(travel_data['hotel_date'])

# 2. Feature engineering - create duration between flight and hotel check-in date (days)
travel_data['days_between_flight_hotel'] = (travel_data['hotel_date'] - travel_data['flight_date']).dt.days

# 3. Encode categorical variables using one-hot encoding or label encoding

# For simplicity, use one-hot encoding for 'flightType', 'agency', 'place', 'company', 'gender'
categorical_cols = ['flightType', 'agency', 'place', 'company', 'gender']

travel_data = pd.get_dummies(travel_data, columns=categorical_cols, drop_first=True)

# 4. Drop columns not useful for modeling like 'travelCode', 'user_name', 'hotel_name'
travel_data.drop(columns=['travelCode', 'user_name', 'hotel_name', 'flight_date', 'hotel_date'], inplace=True)

# 5. Handle any missing values if present (unlikely since previously none found)
travel_data.fillna(method='ffill', inplace=True)

# 6. Optional: Scale numeric columns (like price, distance, age)
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
numeric_cols = ['flight_price', 'time', 'distance', 'hotel_price', 'total', 'days', 'age', 'days_between_flight_hotel']

travel_data[numeric_cols] = scaler.fit_transform(travel_data[numeric_cols])

# Final check
print(travel_data.info())
print(travel_data.head())
````

### 2. Handling Outliers

````python

# STEP 1: IDENTIFY OUTLIERS USING IQR METHOD

print("outlier detection and treatment")

# Select numeric columns for outlier analysis
numeric_cols = travel_data.select_dtypes(include=['float64', 'int64']).columns.tolist()

print("\nNumeric columns to analyze:", numeric_cols)





# METHOD 1: IQR (Interquartile Range) - RECOMMENDED
print("METHOD 1: IQR (Interquartile Range) CAPPING")

travel_data_treated = travel_data.copy()
outlier_summary = []

for col in numeric_cols:
    # Calculate Q1, Q3, and IQR
    Q1 = travel_data[col].quantile(0.25)
    Q3 = travel_data[col].quantile(0.75)
    IQR = Q3 - Q1

    # Define outlier bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Count outliers
    outliers_count = ((travel_data[col] < lower_bound) | (travel_data[col] > upper_bound)).sum()

    outlier_summary.append({
        'Column': col,
        'Q1': Q1,
        'Q3': Q3,
        'IQR': IQR,
        'Lower_Bound': lower_bound,
        'Upper_Bound': upper_bound,
        'Outliers_Count': outliers_count,
        'Outliers_Percentage': (outliers_count / len(travel_data)) * 100
    })

    # Apply IQR capping (replace outliers with bounds)
    travel_data_treated[col] = travel_data_treated[col].clip(lower=lower_bound, upper=upper_bound)

    if outliers_count > 0:
        print(f"\n{col}:")
        print(f"  Q1: {Q1:.2f}, Q3: {Q3:.2f}, IQR: {IQR:.2f}")
        print(f"  Bounds: [{lower_bound:.2f}, {upper_bound:.2f}]")
        print(f"  Outliers found: {outliers_count} ({(outliers_count/len(travel_data)*100):.2f}%)")

# Display outlier summary
print("outlier summary table")
outlier_df = pd.DataFrame(outlier_summary)
print(outlier_df[['Column', 'Outliers_Count', 'Outliers_Percentage']].to_string(index=False))

print("\n IQR-based outlier capping completed!")
print(f" All records preserved: {len(travel_data_treated)} rows")




# METHOD 2: Z-SCORE METHOD (Alternative)
print("METHOD 2: Z-SCORE METHOD (Alternative)")

travel_data_zscore = travel_data.copy()
zscore_outliers = 0

for col in numeric_cols:
    z_scores = np.abs(stats.zscore(travel_data[col].fillna(travel_data[col].mean())))
    outlier_mask = z_scores > 3
    zscore_outliers += outlier_mask.sum()

    if outlier_mask.sum() > 0:
        print(f"{col}: {outlier_mask.sum()} outliers (Z-score > 3)")

print(f"\n Total outliers detected: {zscore_outliers}")
print(f" Note: Z-score method removes rows, resulting in data loss")



# METHOD 3: PERCENTILE CAPPING (Alternative)
print("METHOD 3: PERCENTILE CAPPING (1st-99th percentile)")

travel_data_percentile = travel_data.copy()
percentile_outliers = 0

for col in numeric_cols:
    p1 = travel_data[col].quantile(0.01)
    p99 = travel_data[col].quantile(0.99)

    outliers_count = ((travel_data[col] < p1) | (travel_data[col] > p99)).sum()
    percentile_outliers += outliers_count

    travel_data_percentile[col] = travel_data_percentile[col].clip(lower=p1, upper=p99)

    if outliers_count > 0:
        print(f"{col}: {outliers_count} outliers capped (P1: {p1:.2f}, P99: {p99:.2f})")

print(f"\n Total outliers capped: {percentile_outliers}")



# STEP 2: COMPARISON OF METHODS

print("comparison: method effectiveness")

comparison = pd.DataFrame({
    'Method': ['IQR Capping', 'Z-Score', 'Percentile Capping'],
    'Rows Preserved': [len(travel_data_treated), len(travel_data_zscore), len(travel_data_percentile)],
    'Data Loss': [0, len(travel_data) - len(travel_data_zscore), 0],
    'Suitable_For': ['General Use', 'Extreme Outliers', 'Skewed Data']
})

print("\n", comparison.to_string(index=False))





# STEP 3: VISUALIZE BEFORE AND AFTER
print("VISUALIZATION: Before & After Outlier Treatment")

# Visualize for key columns
key_cols = ['flight_price', 'hotel_price', 'age', 'distance']
key_cols = [col for col in key_cols if col in travel_data.columns]

fig, axes = plt.subplots(2, len(key_cols), figsize=(16, 8))

for idx, col in enumerate(key_cols):
    # Before treatment - Box plot
    axes[0, idx].boxplot(travel_data[col].dropna())
    axes[0, idx].set_title(f'{col}\n(BEFORE Treatment)', fontweight='bold')
    axes[0, idx].set_ylabel('Value')
    axes[0, idx].grid(True, alpha=0.3)

    # After treatment - Box plot
    axes[1, idx].boxplot(travel_data_treated[col].dropna())
    axes[1, idx].set_title(f'{col}\n(AFTER IQR Treatment)', fontweight='bold')
    axes[1, idx].set_ylabel('Value')
    axes[1, idx].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("Box plots generated for visual comparison")




# STEP 4: FINAL DATASET AFTER TREATMENT
print("FINAL DATASET AFTER OUTLIER TREATMENT")

print(f"\nOriginal dataset shape: {travel_data.shape}")
print(f"After IQR treatment: {travel_data_treated.shape}")
print(f"Data preservation rate: {(len(travel_data_treated)/len(travel_data)*100):.2f}%")




# Use treated data for next steps
travel_data = travel_data_treated

print("\n Outlier treatment completed successfully!")
print("  Dataset ready for categorical encoding and feature scaling!")
````

##### What all outlier treatment techniques have you used and why did you use those techniques?

Answer Here.


I used three main outlier treatment methods, each selected for its advantages in handling travel data:

1. **IQR (Interquartile Range) Capping:**  
   - This method identifies outliers as values outside 1.5 times the IQR below Q1 or above Q3 and caps them to those boundary limits.  
   - **Why?** It effectively limits the impact of extreme values while preserving all data points, maintaining realistic price and distance variations common in travel datasets. It prevents loss of valuable records and is widely accepted for business use.

2. **Z-Score Based Removal:**  
   - Outliers are detected as values with a Z-score (number of standard deviations from mean) greater than 3, and corresponding rows are removed.  
   - **Why?** This method removes extreme, rare anomalies that can negatively affect model training but leads to data loss, so itâ€™s best used when dataset size is large or outliers are clearly errors.

3. **Percentile Capping (Winsorization):**  
   - Caps values that fall below the 1st percentile or above the 99th percentile at those limits.

#### 1. Feature Manipulation

````python
# Manipulate Features to minimize feature correlation and create new features




# step 1: analyze existing numeric features

print("STEP 1: Existing Numeric Features Analysis")

numeric_cols = travel_data.select_dtypes(include=['float64', 'int64']).columns.tolist()
print(f"\n Original numeric features ({len(numeric_cols)}): {numeric_cols}")
print(f"Dataset shape before manipulation: {travel_data.shape}")

# step 2: create new features - interaction features

print("STEP 2: create new features - interaction & derived features")

travel_data = travel_data.copy()
new_features = []




# Feature 1: Price per Distance (efficiency metric)
if 'flight_price' in travel_data.columns and 'distance' in travel_data.columns:
    travel_data['flight_price_per_km'] = (
        travel_data['flight_price'] / (travel_data['distance'] + 1)
    )
    new_features.append('flight_price_per_km')
    print("\n Created: flight_price_per_km")
    print(f"  Logic: flight_price / (distance + 1)")
    print(f"  Interpretation: Cost per unit distance traveled")



# Feature 2: Hotel Price per Day (daily rate)
if 'hotel_price' in travel_data.columns and 'days' in travel_data.columns:
    travel_data['hotel_price_per_day'] = (
        travel_data['hotel_price'] / (travel_data['days'] + 1)
    )
    new_features.append('hotel_price_per_day')
    print("\n Created: hotel_price_per_day")
    print(f"  Logic: hotel_price / (days + 1)")
    print(f"  Interpretation: Daily hotel accommodation cost")





# Feature 3: Total Travel Cost (sum of major costs)
if 'flight_price' in travel_data.columns and 'hotel_price' in travel_data.columns:
    travel_data['total_accommodation_cost'] = (
        travel_data['flight_price'] + travel_data['hotel_price']
    )
    new_features.append('total_accommodation_cost')
    print("\n Created: total_accommodation_cost")
    print(f"  Logic: flight_price + hotel_price")
    print(f"  Interpretation: Combined major travel expenses")





# Feature 4: Hotel to Flight Cost Ratio
if 'hotel_price' in travel_data.columns and 'flight_price' in travel_data.columns:
    travel_data['hotel_flight_cost_ratio'] = (
        travel_data['hotel_price'] / (travel_data['flight_price'] + 1)
    )
    new_features.append('hotel_flight_cost_ratio')
    print("\n Created: hotel_flight_cost_ratio")
    print(f"  Logic: hotel_price / (flight_price + 1)")
    print(f"  Interpretation: Proportion of cost on accommodation vs. flight")





# Feature 5: Distance per Day (travel intensity)
if 'distance' in travel_data.columns and 'days' in travel_data.columns:
    travel_data['distance_per_day'] = (
        travel_data['distance'] / (travel_data['days'] + 1)
    )
    new_features.append('distance_per_day')
    print("\n Created: distance_per_day")
    print(f"  Logic: distance / (days + 1)")
    print(f"  Interpretation: Travel intensity (distance covered per day)")





# Feature 6: Age Group Buckets (categorical grouping)
if 'age' in travel_data.columns:
    bins = [0, 18, 25, 35, 50, 65, 100]
    labels = ['Child', 'Young Adult', 'Adult', 'Middle-aged', 'Senior', 'Elderly']
    travel_data['age_group'] = pd.cut(travel_data['age'], bins=bins, labels=labels, include_lowest=True)
    new_features.append('age_group')
    print("\n Created: age_group (categorical)")
    print(f"  Bins: {bins}")
    print(f"  Labels: {labels}")






# Feature 7: Trip Duration Category
if 'days' in travel_data.columns:
    bins = [0, 3, 7, 14, 30, 365]
    labels = ['Weekend', 'Short', 'Week', 'Long', 'Extended']
    travel_data['trip_duration_category'] = pd.cut(travel_data['days'], bins=bins, labels=labels, include_lowest=True)
    new_features.append('trip_duration_category')
    print("\n Created: trip_duration_category (categorical)")
    print(f"  Categories: {labels}")






# Feature 8: Cost per Person per Day (if user_count available)
if all(col in travel_data.columns for col in ['total', 'days', 'user_count']):
    travel_data['cost_per_person_per_day'] = (
        travel_data['total'] / ((travel_data['days'] + 1) * (travel_data['user_count'] + 1))
    )
    new_features.append('cost_per_person_per_day')
    print("\n Created: cost_per_person_per_day")
    print(f"  Logic: total / ((days + 1) * (user_count + 1))")
    print(f"  Interpretation: Per-person daily travel cost")





# Feature 9: Is Long Distance Travel (binary)
if 'distance' in travel_data.columns:
    distance_median = travel_data['distance'].median()
    travel_data['is_long_distance'] = (travel_data['distance'] > distance_median).astype(int)
    new_features.append('is_long_distance')
    print("\n Created: is_long_distance (binary)")
    print(f"  Threshold: median distance = {distance_median:.2f}")





# Feature 10: Is Premium Travel (high cost indicator)
if 'total' in travel_data.columns:
    total_75th = travel_data['total'].quantile(0.75)
    travel_data['is_premium_travel'] = (travel_data['total'] > total_75th).astype(int)
    new_features.append('is_premium_travel')
    print("\nâœ“ Created: is_premium_travel (binary)")
    print(f"  Threshold: 75th percentile total = {total_75th:.2f}")

print(f"\n Total new features created: {len(new_features)}")






# step 3: final summary

print("STEP 3: FEATURE MANIPULATION SUMMARY")

print(f"\nBefore feature manipulation: {len(numeric_cols)} numeric features")
print(f"New features created: {len(new_features)}")
print(f"Dataset shape before: (original)")
print(f"Dataset shape after: {travel_data.shape}")

print(f"\nAll features in dataset:")
all_features = travel_data.columns.tolist()
for i, col in enumerate(all_features, 1):
    print(f"  {i}. {col}")

print("\n Feature manipulation completed successfully")
print(" All new features created and added to dataset")
print("\n  Dataset ready for Feature Selection")
````

````python
# ========================================
# âœ… Final SVR Model with Derived Features (Fixed)
# ========================================

import pandas as pd
import numpy as np
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split, RandomizedSearchCV, KFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from scipy.stats import uniform
import joblib

# Step 1: Load dataset
travel_data = pd.read_csv('/content/travel_capstone/travel_data.csv')
print("âœ… Dataset loaded successfully!")
print("Shape:", travel_data.shape)

# Step 2: Create derived features
travel_data['hotel_price_per_day'] = travel_data['hotel_price'] / travel_data['days']
travel_data['distance_per_day'] = travel_data['distance'] / travel_data['days']
travel_data['travel_intensity'] = travel_data['distance'] * travel_data['time']

# Step 3: Feature selection
selected_features = [
    'flight_price',
    'hotel_price',
    'days',
    'distance',
    'hotel_price_per_day',
    'distance_per_day',
    'travel_intensity'
]
X = travel_data[selected_features]
y = travel_data['total']

# Step 4: Scale data
x_scaler = StandardScaler()
y_scaler = StandardScaler()
X_scaled = x_scaler.fit_transform(X)
y_scaled = y_scaler.fit_transform(y.values.reshape(-1, 1)).ravel()

# Step 5: Split data
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y_scaled, test_size=0.2, random_state=42
)
print(f"âœ… Train-Test Split Done â†’ Train: {X_train.shape[0]}, Test: {X_test.shape[0]}")

# Step 6: SVR with RandomizedSearchCV
param_dist = {
    'kernel': ['rbf'],
    'C': uniform(0.1, 10),
    'epsilon': uniform(0.01, 0.2)
}

cv = KFold(n_splits=3, shuffle=True, random_state=42)
random_search = RandomizedSearchCV(
    estimator=SVR(),
    param_distributions=param_dist,
    n_iter=5,
    cv=cv,
    scoring='neg_mean_squared_error',
    n_jobs=-1,
    random_state=42,
    verbose=1
)

random_search.fit(X_train, y_train)
best_svr = random_search.best_estimator_

# Step 7: Evaluate
y_train_pred = best_svr.predict(X_train)
y_test_pred = best_svr.predict(X_test)

y_train_real = y_scaler.inverse_transform(y_train.reshape(-1, 1))
y_test_real = y_scaler.inverse_transform(y_test.reshape(-1, 1))
y_train_pred_real = y_scaler.inverse_transform(y_train_pred.reshape(-1, 1))
y_test_pred_real = y_scaler.inverse_transform(y_test_pred.reshape(-1, 1))

train_mae = mean_absolute_error(y_train_real, y_train_pred_real)
test_mae = mean_absolute_error(y_test_real, y_test_pred_real)
train_r2 = r2_score(y_train_real, y_train_pred_real)
test_r2 = r2_score(y_test_real, y_test_pred_real)

print("\nðŸ“Š === Model Evaluation (in â‚¹) ===")
print(f"Train â†’ MAE: â‚¹{train_mae:.2f}, RÂ²: {train_r2:.4f}")
print(f"Test  â†’ MAE: â‚¹{test_mae:.2f}, RÂ²: {test_r2:.4f}")
print(f"Best Parameters: {random_search.best_params_}")

# Step 8: Save model and scalers
joblib.dump(best_svr, 'best_svr_model.joblib')
joblib.dump(x_scaler, 'x_scaler.joblib')
joblib.dump(y_scaler, 'y_scaler.joblib')
print("âœ… Model and scalers saved successfully!")

# Step 9: Sanity Check (Different Inputs)
test_samples = [
    [4500, 2200, 5, 800, 440, 160, 1800],   # expensive trip
    [1000, 500, 2, 400, 250, 125, 800],     # medium trip
    [500, 200, 1, 250, 200, 200, 400]       # budget trip
]

for i, sample in enumerate(test_samples, 1):
    scaled_input = x_scaler.transform([sample])
    pred_scaled = best_svr.predict(scaled_input)
    pred_real = y_scaler.inverse_transform(pred_scaled.reshape(-1, 1))
    print(f"ðŸ§¾ Sample {i} â†’ Predicted Total Cost: â‚¹{pred_real[0][0]:.2f}")

````

### 3. Explain the model which you have used and the feature importance using any model explainability tool?

### Model Explanation:

- **SVR** aims to find a function that deviates from the actual target values by a value no greater than epsilon (Îµ) for all training points and is as flat as possible. It uses kernel functions (like RBF used here) to handle non-linear relationships by mapping input features into high-dimensional space.
- SVR maintains a balance between model complexity and prediction accuracy by controlling the margin and penalizing deviations outside the epsilon margin.
- It is effective for datasets with non-linear relationships and moderate feature dimensions.

### Feature Importance and Explainability:

- SVR does not directly provide explicit feature importance scores like tree-based models.
- To interpret SVR, **model-agnostic explainability tools** such as **SHAP (SHapley Additive exPlanations)** or **LIME (Local Interpretable Model-Agnostic Explanations)** are used.
- These tools explain the impact of each feature on individual predictions and globally on the model behavior by approximating the underlying model locally or through cooperative game theory principles.
- SHAP values provide additive feature importance that can be visualized through summary plots, dependence plots, and force plots, offering insight into how each feature contributes positively or negatively to predictions.

Using such model explainability tools ensures transparency and trust in the SVR model, enabling stakeholders to understand which features drive predictions and make informed decisions accordingly.
